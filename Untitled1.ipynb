### Data Extraction/Scrapping using Beautiful Soup

### steps-
<ul>
<li>Import Beautiful Soup
<li>Make a Get request to the web page using Request Library(requests or urllib)
<li>Parse HTML
<li>Filter relevant parts
<li>storing info in a csv(comma seperated file) file</ul>

## step1

```python
# import Library bs4 package which contain beautiful soup class
import bs4


# import request library to make a Get request -  a get request is a request to get the desired web page's 
# source code to us. we are client and we want data from internet (server) so we make a get request and get a html 
# response. after that we will parse the data to get desired data

from urllib.request import urlopen

url="https://en.wikipedia.org/wiki/Android_version_history"
data=urlopen(url)

print(type(data)) #http response objects
```

```python
html_source_code= data.read()
print(html_source_code)

data.close()
#closing connection after getting data
```

## step - Parsing 

## what is parsing
 example - parsing json: JSON is a format specification as mentioned by the rest. Parsing JSON means interpreting the data with whatever language u are using at the moment

```python
from bs4 import BeautifulSoup as BS

soup_class=BS(html_source_code,'html.parser')
#2nd parameter is tell the type of document getting parsed- json,html etc

print(soup_class)
#we didn't get an exact source code as we write an html doc from html response but after parsing that we got 
#exact html
```

```python
print(soup_class.h1) #first h1 tage in html
print(soup_class.find_all('h1',{})) #print all h1 tags
```

```python
all_tables=soup_class.find_all('table',{'class':'wikitable'})
print(all_tables)
```

```python
desired_table=all_tables[0]
```

## step - Extracting useful information
<ul>
<li>Removing undesired tags
<li>Extract table header(column names) and data
</ul>

```python
headers=desired_table.find_all('th')
headers=[h.text for h in headers]
print(headers)

column_titles=[h[:-1] for h in headers]
print(headers)
```

```python
row_data=desired_table.find_all('tr')[1:] #slicing because 1st row will be of headers
row_data=[[w.text[:-1] for w in r.find_all('td')] for r in row_data]

print(row_data)
```

## step - writing and reading CSV files

```python
filename="android_version_history.csv"
with open(filename,'w') as f:
    header_string=','.join(column_titles)
    header_string+='\n'
    f.write(header_string)
    
    for r in row_data:
        row_string=','.join(r)
        row_string+='\n'
        f.write(row_string)
```
