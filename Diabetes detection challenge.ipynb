```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing

plt.style.use('seaborn')
```

```python
dfx=pd.read_csv('Datasets/Diabetes/TrainingData/Diabetes_XTrain.csv')
dfy=pd.read_csv('Datasets/Diabetes/TrainingData/Diabetes_YTrain.csv')
dfx.head()
```

```python
data=pd.concat([dfx,dfy],axis=1)
correlation_matrix=data.corr()
correlation_matrix
```

```python
dfx=dfx.drop(columns=['BloodPressure','SkinThickness','Insulin','DiabetesPedigreeFunction'])
x=dfx.values
y=dfy.values
print(x.shape,y.shape)
```

```python
def dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
def KNN(X,Y,query,k=13):
    distance=[]
    size=X.shape[0]
    for i in range(size):
        d=dist(X[i],query)
        distance.append([d,Y[i]])
    distance.sort()
    distance=distance[:k]
    distance=np.array(distance)
    distance=np.unique(distance[:,1],return_counts=True)
    idx=distance[1].argmax()
#     print(distance,idx,distance[idx])
    return distance[0][idx]
```

```python
df=pd.read_csv('Datasets/Diabetes/Diabetes_Xtest.csv')
df=df.drop(columns=['BloodPressure','SkinThickness','Insulin','DiabetesPedigreeFunction'])
xtest=df.values
print(xtest.shape)
ypred=[]
for i in range(xtest.shape[0]):
    pred=int(KNN(x,y,xtest[i]))
    ypred.append(pred)
```

```python
dfpred=pd.DataFrame(ypred,dtype=np.int64)
dfpred.to_csv('solution.csv',index=False,header=['Outcome'])
```
