```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras import models
from keras.layers import Dense
```

```python
df=pd.read_csv('Datasets/House Price Prediction NN/Train/Train_Data.csv')
dftest=pd.read_csv('Datasets/House Price Prediction NN/Test/Test_Data.csv')

label=dftest['Id'].values
df = df.drop(['Id','Street','LandContour', 'Utilities', 'LandSlope','Condition1','Condition2',
              'RoofMatl','Heating','CentralAir','Electrical','Functional','GarageQual',
              'GarageCond','PavedDrive', 'SaleType', 'YearBuilt', 'GarageYrBlt', 
              'GarageArea', 'PoolArea', 'MoSold', 'SaleCondition'], axis=1)

dftest = dftest.drop(['Id','Street','LandContour', 'Utilities', 'LandSlope','Condition1','Condition2',
              'RoofMatl','Heating','CentralAir','Electrical','Functional','GarageQual',
              'GarageCond','PavedDrive', 'SaleType', 'YearBuilt', 'GarageYrBlt', 
              'GarageArea', 'PoolArea', 'MoSold', 'SaleCondition'], axis=1)

df.head()
```

```python
# df.info()
```

# Train Data

```python

# dropping missing columns and data that doesn't tell much
null_values_per_col = np.sum(df.isnull(), axis=0)
max_NAN = int(2*df.shape[0]/3.0)
columns_to_drop=[]
columns=df.columns.to_list()
for col in columns:
    if null_values_per_col[col]>max_NAN:
        columns_to_drop.append(col)
df=df.drop(columns_to_drop,axis=1)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer =imputer.fit(df)
df=pd.DataFrame(imputer.transform(df),columns=df.columns.to_list()) #return numpy array 
print(df.shape)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns.to_list():
    if(df[col].dtype == "object"):
        df[col] = le.fit_transform(df[col])
```

# Test Data

```python

null_values_per_col = np.sum(dftest.isnull(), axis=0)
max_NAN = int(2*dftest.shape[0]/3.0)
columns_to_drop=[]
columns=dftest.columns.to_list()
for col in columns:
    if null_values_per_col[col]>max_NAN:
        columns_to_drop.append(col)
dftest=dftest.drop(columns_to_drop,axis=1)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

imputer =imputer.fit(dftest)
dftest=pd.DataFrame(imputer.transform(dftest),columns=dftest.columns.to_list()) #return numpy array 
dftest.head()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in dftest.columns.to_list():
    if(dftest[col].dtype == "object"):
        dftest[col] = le.fit_transform(dftest[col])
```

```python
xtrain=df.values[:,:-1]
ytrain=df.values[:,-1]
xtest=dftest.values

# from sklearn.preprocessing import StandardScaler
# sc_X = StandardScaler()
# xtrain = sc_X.fit_transform(xtrain)
# xtest=sc_X.transform(xtest)
```

```python
# #Neural Network
model=models.Sequential()
model.add(Dense(800,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dense(600,activation='relu'))
model.add(Dense(400,activation='relu'))
model.add(Dense(200,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(1,activation='linear'))
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])
```

```python
# model.summary()
```

```python
ytrain=ytrain.reshape((ytrain.shape[0],1))
print(xtrain.shape,ytrain.shape,type(ytrain))
xtrain=np.asarray(xtrain).astype(np.float32)
ytrain=np.asarray(ytrain).astype(np.float32)
h=model.fit(xtrain,ytrain,batch_size=256,epochs=200)
```

```python
h=model.evaluate(xtrain,ytrain)
print(h)
```

```python
y=model.predict(xtest)
label=label.reshape((-1,1))
y=np.hstack((label,y))
dfy=pd.DataFrame(y,columns=['Id','Salesprice'])
dfy.astype({'Id':'int32'})
```
