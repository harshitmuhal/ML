```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras import models
from keras.layers import Dense
```

```python
x=pd.read_csv('Datasets/classifypoints/Training Data/Logistic_X_Train.csv').values
y=pd.read_csv('Datasets/classifypoints/Training Data/Logistic_Y_Train.csv').values
print(x.shape,y.shape)
```

```python
#Neural Network
model=models.Sequential()
model.add(Dense(16,activation='relu',input_shape=(2,))) #2 features
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
h=model.fit(x,y,batch_size=128,validation_split=0.3,epochs=140)
```

```python
xtest=pd.read_csv('Datasets/classifypoints/Test Cases/Logistic_X_Test.csv').values
y=model.predict(xtest)
```

```python
ans=np.zeros(y.shape)
ans[y>0.5]=1
ans=ans.astype('int32')
```

```python
dfy=pd.DataFrame(ans)
dfy.head()
```

```python
dfy.to_csv('classifypoints_solution.csv',index=False)
```
