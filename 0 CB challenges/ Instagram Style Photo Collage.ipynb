```python
import cv2
```

```python
top_left=cv2.imread('../Datasets/ImageCollage/top_left.jpg')
top_right=cv2.imread('../Datasets/ImageCollage/top_right.jpg')
bottom_left=cv2.imread('../Datasets/ImageCollage/bottom_left.jpg')
bottom_right=cv2.imread('../Datasets/ImageCollage/bottom_right.jpg')
```

```python
center=cv2.imread('../Datasets/ImageCollage/center.jpeg')
```

```python
center=cv2.resize(center,(100,100))
```

```python
top_left=cv2.resize(top_left,(200,200))
top_right=cv2.resize(top_right,(200,200))
bottom_left=cv2.resize(bottom_left,(200,200))
bottom_right=cv2.resize(bottom_right,(200,200))
```

```python
top_left=cv2.cvtColor(top_left, cv2.COLOR_BGR2RGB)
top_right=cv2.cvtColor(top_right, cv2.COLOR_BGR2RGB)
bottom_left=cv2.cvtColor(bottom_left, cv2.COLOR_BGR2RGB)
bottom_right=cv2.cvtColor(bottom_right, cv2.COLOR_BGR2RGB)
center=cv2.cvtColor(center, cv2.COLOR_BGR2RGB)
```

```python
import numpy as np
img=np.zeros((430,430,3))
```

```python
temp=np.zeros((120,120,3))
temp[10:110,10:110,]=center
print(temp.shape)
```

```python
img[10:210,10:210,:]=top_left
img[10:210,220:420,:]=top_right
img[220:420,10:210,:]=bottom_left
img[220:420,220:420,:]=bottom_right
img[155:275,155:275,:]=temp
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.axis('off')
n_img=np.array(img,dtype='int32')
# print(n_img)
plt.imshow(n_img)
plt.show()
```

```python
print(img.shape)
img=np.array(img,dtype='int32')
img.resize((img.shape[0]*img.shape[1],3))
```

```python
import pandas as pd
df=pd.DataFrame(img,columns=['r','g','b'])
```
