```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('seaborn')
```

```python
dfx=pd.read_csv('Datasets/xdata.csv')
dfy=pd.read_csv('Datasets/ydata.csv')
X=dfx.values[:,1:]
Y=dfy.values[:,1:].reshape((-1,)) #-1 means preserve same rows 
#reshaping is required bcz in scatter function c take a 1D array
print(X.shape,Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)  #to see how c works check - plt.scatter? documentation
query_x = np.array([0,1])
plt.scatter(query_x[0],query_x[1],color='red')
plt.show()
```

```python
def euclidian_dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
def knn(X,Y,query,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=euclidian_dist(X[i],query)
        vals.append([d,Y[i]])
    vals.sort()
    vals=vals[:k]
    vals=np.array(vals)
    vals=np.unique(vals[:,1],return_counts=True)
#     print(vals)
    idx=vals[1].argmax()
    return vals[idx][0]
```

```python
x=knn(X,Y,query_x,5)
print(x)
```
