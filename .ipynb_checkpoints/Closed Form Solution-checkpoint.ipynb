#### Closed Form Solution in Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate Dataset
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
Y = Y.reshape((-1,1))
print(X.shape)
print(Y.shape)
```

```python
# Normalize
X = (X-X.mean())/X.std()
# Visualize
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_.shape)
print(X_[:5,:])
```

```python
def predict(X,theta):
    return np.dot(X,theta)
def getThetaClosedForm(X,Y):
    Y = np.mat(Y)
    firstPart = np.dot(X.T,X)
    secondPart = np.dot(X.T,Y)
    theta = np.linalg.pinv(firstPart)*secondPart #pseudoinverse from linear algebra model
    #pseudo inverse is equal to multiplicative inverse if inverse exists. if not then it
    #gives an approximate inverse
    return theta
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label="prediction")
plt.title("Normalized Data")
plt.legend()
plt.show()
```
