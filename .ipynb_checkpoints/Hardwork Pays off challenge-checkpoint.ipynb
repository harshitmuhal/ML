```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('seaborn')
```

```python
X=pd.read_csv('Datasets/hardworkpay_challenge/TrainingData/Linear_X_Train.csv').values
Y=pd.read_csv('Datasets/hardworkpay_challenge/TrainingData/Linear_Y_Train.csv').values

print(X.shape,Y.shape)

plt.scatter(X,Y)
plt.show()
```

```python
def hypothesis(X,theta):
    ypred=theta[1]*X+theta[0]
    return ypred
def gradient(X,Y,theta):
    grad=np.zeros((2,))
    ypred=hypothesis(X,theta)
    m=X.shape[0]
    grad[0]=np.sum(ypred-Y)
    grad[1]=np.sum((ypred-Y)*X)
    return grad/m
def mean_squared_error(X,Y,theta):
    ypred=hypothesis(X,theta)
    return sum((ypred-Y)**2)/X.shape[0]
def gradientDescent(X,Y,theta,learning_rate=0.1,steps=100):
    error=[]
    for i in range(steps):
        grad=gradient(X,Y,theta)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        error.append(mean_squared_error(X,Y,theta))
    return error,theta
def r2_score(Y,Y_):
    num = np.sum((Y-Y_)**2)
    denom = np.sum((Y- Y.mean())**2)
    score = (1- num/denom)
    return score*100
```

```python
theta=np.zeros((2,))
error,theta = gradientDescent(X,Y,theta)
print(theta)
plt.plot(error)
plt.show()
```

```python
ypred=hypothesis(X,theta)
print(r2_score(Y,ypred))
plt.scatter(X,Y)
plt.plot(X,ypred,color='orange')
plt.show()
```

```python
x=pd.read_csv('Datasets/hardworkpay_challenge/TestCases/Linear_X_Test.csv').values
ytest=hypothesis(x,theta)
df=pd.DataFrame(ytest,columns=['y'])
df.to_csv('ypred.csv',index=False)
```
