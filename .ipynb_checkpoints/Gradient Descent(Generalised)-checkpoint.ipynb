# Gradient Descent - 
For given function F(x) , Find value of x at which F(x) is minimum

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10)
Y=(X-5)**2
```

```python
#method 1

x=0 #initialse x with 0
step=0.1
plt.plot(X,Y)
plt.title('Gradient Descent Visualisation')
plt.xlabel('x')
plt.ylabel('F(x)')
for i in range(50):
    grad=2*(x-5)
    x=x-step*grad
    y=(x-5)**2
    plt.scatter(x,y)
plt.show()
print(x)

#see that as we move closer to the required point effective size decreases because 
#slope/gradient decreases as we are moving towards the local minima
```

```python
#method 1

x=0 #initialse x with 0
step=0.1
plt.plot(X,Y)
plt.title('Gradient Descent Visualisation')
plt.xlabel('x')
plt.ylabel('F(x)')
while 1:
    grad=2*(x-5)
    delta=x
    x=x-step*grad
    delta-=x
    delta=abs(delta)
    y=(x-5)**2
    plt.scatter(x,y)
    if(delta<0.00001):
        break
plt.show()
print(x)
```
