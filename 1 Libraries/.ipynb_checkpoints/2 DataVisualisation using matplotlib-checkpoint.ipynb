```python
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available)
plt.style.use('seaborn')
```

# line plot 

```python
x=np.arange(10)
y1=x**3/2
y2=x**2
y3=20*x+7
```

```python
plt.plot(x,y1,color='red',label='line y=x^3/2')
#marker puts dot at (x,y) values used to make graph
plt.plot(x,y2,marker='o',label='line y=x^2') 
plt.xlabel('ye mera x label hai')
plt.ylabel('ye mera y label hai')
plt.title('practise graph')
#legend() function is used to display labels for each line
plt.legend()
plt.show()

plt.plot(x,y3,marker='^',label='graph',color='black',linestyle='dashed')
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(4,4))

plt.scatter(x,y1,color='red',label='line y=x^3/2')
#marker puts dot at (x,y) values used to make graph
plt.scatter(x,y2,marker='o',label='line y=x^2') 
plt.xlabel('ye mera x label hai')
plt.ylabel('ye mera y label hai')
plt.title('practise graph')
#legend() function is used to display labels for each line
plt.legend()
plt.show()

plt.figure(figsize=(4,4))
plt.scatter(x,y3,marker='^',label='graph',color='black')
plt.show()
```

# Bar Graphs

```python
plt.bar?
```

```python
x=np.array([1,3,5])
height1=np.array([10,5,15])
height2=np.array([8,5,14])

w=0.50
#x-w/2 and x+w/2 is trick to avoid overlapping
plt.figure(figsize=(4,4))
plt.bar(x-w/2,height1,width=w,color='orange',label='2019',tick_label=['a','b','c'])
plt.bar(x+w/2,height2,width=w,color='red',label='2019',tick_label=['a','b','c'])
plt.xlabel('ye mera x label hai')
plt.ylabel('ye mera y label hai')
plt.title('practise graph')
plt.xlim(0,6)
plt.ylim(0,20)
plt.legend()
plt.show()
```

# pie chart

```python
plt.pie?
```

```python
labels = ['Algebra','Calculus','Coding','Algorithms']
values = [90,80,54,40]

plt.pie(values,labels=labels)
plt.show()

plt.pie(values,labels=labels,explode=(0,0.2,0,0),shadow=True,radius=0.5)
plt.show()
```

# Histogram

```python
# Generate a Normal Distribution (1-D)
u = 5
sigma = 0.2
# Such a distribution is also called standard normal distribution
vals = u + sigma*np.random.randn(200)
print(vals.shape)
plt.figure(figsize=(4,4))
plt.hist(vals,10)
plt.show()
```
